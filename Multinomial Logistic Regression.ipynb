```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
data =load_iris()
data
```

```python
df =pd.DataFrame(data.data ,columns=data.feature_names)
```

```python
df["target"] =data.target
```

```python
df
```

```python
df.target.value_counts()
```

```python
df
```

#### Seprate  the X and Y columns

```python
X =df.iloc[:,:-1]
y =df.iloc[:,-1]
```

```python
X
```

```python
y
```

#### Train the model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.25,random_state=12)
```

```python
X_train.shape ,X_test.shape
```

#### Build the multinomial model

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(X_train ,y_train)
y_pred =model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix ,accuracy_score,classification_report
print(confusion_matrix(y_test ,y_pred))
```

```python
print(accuracy_score(y_test ,y_pred))
```

```python
print(classification_report(y_test ,y_pred))
```

#### Check the best parameter using grid search CV 

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.model_selection import GridSearchCV
# classifier =LogisticRegression()
param ={"penalty":("l1","l2","elasticnet"),"C":[1,2,3,4,20,30,50],"solver":['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']}
clf =GridSearchCV(model ,param_grid=param,verbose=3)
clf.fit(X_train ,y_train)
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
r_clf =RandomizedSearchCV(model, param_distributions=param ,verbose=3)
r_clf.fit(X_train ,y_train)
```

```python
r_clf.best_estimator_
```

```python
r_clf.best_params_
```

```python
r_clf.best_score_
```
